```python
import sys

with open('C:/bestsellers.txt') as file:
    lines = file.readlines()

books = []
for x in range(len(lines)):
    book = list(lines[x].split('\t'))
    for y in range(len(book)):
        book[y] = book[y].strip()
    books.append(book)

def holdon_and_continue():
    '''
    None->None
    This function allows the user to hit enter, then calls main().
    '''
    print('\n'*2)
    a = input('Hit the Enter key to continue...')
    print('\n'*2)
    main()

def year_range(fyear, lyear):
    '''
    (Int, Int)
    This takes as inputs of two integers of length 4, and searches the book for the
    year range.
    '''
    years = []
    book_list = []
    for year in range(fyear, lyear+1):
        years.append(str(year))
    
    for y in years:
        for x in range(len(books)):
            if y in books[x][3]:
                book_list.append(books[x][0]+' by '+books[x][1]+', ('+books[x][3]+')')
    if(len(book_list) == 0):
        print('No books found.')
    else:
        for z in book_list:
            print(z)
    holdon_and_continue()

def month_year(month, year):
    '''
    (int, int)
    This takes input of a month integer of length 2, where 1 <= month <= 12, and a
    year integer of 4 digits in length.
    '''
    book_list = []
    for x in range(len(books)):
        slash_index = books[x][3].index('/')
        if((str(year) in books[x][3]) and \
           (books[x][3][0:slash_index] == str(month))):
            book_list.append(books[x])
    print('\n')
    
    if(len(book_list) < 1):
        print('No books found.')
    else:
        for y in range(len(book_list)):
            print(book_list[y][0],'by',book_list[y][1]+', ('+book_list[y][3]+')')
    holdon_and_continue()

def search_author(author):
    '''
    Prompts the user for a string and displays all books that contains
    string in the author's name.
    '''
    author = author.lower()
    book_list = []
    
    for x in range(len(books)):
        if(author in books[x][1].lower()):
            book_list.append(books[x])
    if(len(book_list) == 0):
        print('No books found.')
    else:
        for y in range(len(book_list)):
            print(book_list[y][0],'by',book_list[y][1]+', ('+book_list[y][3]+')')
    holdon_and_continue()

def search_title(title):
    '''
    Prompt the user for a string that displays books that contains
    string in the book's title.
    '''
    title = title.lower()
    book_list = []
    
    for x in range(len(books)):
        if(title in books[x][0].lower()):
            book_list.append(books[x])
    if(len(book_list) == 0):
        print('No books found.')
    else:
        for y in range(len(book_list)):
            print(book_list[y][0],'by',book_list[y][1]+', ('+book_list[y][3]+')')
    holdon_and_continue()

def main():
    '''
    None->None
    Main Function.
    '''
    print('Enter 1, 2, 3, 4, or Q for answer.')
    print('1. Book in year range')
    print('2. Book in month/year')
    print('3. Search for author in selected string')
    print('4. Search for title in selected string')
    print('Q: Quit')
    
    running = True
    
    while running:
        user_input = input('Answer (1,2,3,4,Q or q):')
        user_input = user_input.lower()
        if(user_input == 'q'):
            print('Exiting the program')
            print('Bye!')
            sys.exit(0)
        elif(user_input == '1'):
            print('Search year range:')
            valid1 = False
            valid2 = False
            while valid1 is False:
                try:
                    y1 = int(input('Enter the first year:\n'))
                    valid1 = True
                except ValueError:
                    print('Must be a four digit number.')
            if((y1 // 1000) < 1):
                valid1 = False
                print('Must be four digits, e.g., 1981 or 2019.')
            else:
                valid1 = True
                pass
            
            while valid2 is False:
                try:
                    y2 = int(input('Enter the last year:\n'))
                    valid2 = True
                except ValueError:
                    print('Must be a four digit number.')
            if((y2 // 1000) < 1):
                valid2 = False
                print('Must be four digits, 1981 or 2019.')
            else:
                valid2 = True
                pass
            running = False
            year_range(y1,y2)      
        elif(user_input == '2'):
            print('Search month/year')
            valid1 = False
            valid2 = False
            while valid1 is False:
                try:
                    mm = int(input('Enter month (as an integer 1-12): '))
                    if(mm < 1) or (mm > 12):
                        print('Month must be between 1 and 12.')
                        valid1 = False
                    else:
                        valid1 = True
                except ValueError:
                    print('Month must be integer between 1-12.')
            
            while valid2 is False:
                try:
                    year = int(input('Enter year: '))
                    if((year // 1000) < 1):
                        print('Year must be integer with 4 digits.')
                        valid2 = False
                    else:
                        valid2 = True
                except ValueError:
                    print('Year must be integer with 4 digits.')
            
            running = False
            month_year(mm, year)
            holdon_and_continue()
        elif(user_input == '3'):
            author = input('Enter a string as part of the authors name: ')
            search_author(author)
        elif(user_input == '4'):
            title = input('Enter a string as part of the book title: ')
            search_title(title)
        else:
            print('That is an invalid command!')
if __name__ == '__main__':
    main()
```
